```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
```

```python
# Generate Dataset
X,Y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=7.8,random_state=11)
print(X.shape)
print(Y.shape)
Y =Y.reshape((-1,1))
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X,Y)
```

```python
# Normalize
X = (X-X.mean())/X.std()

# Visualize
plt.figure()
plt.scatter(X,Y)
plt.title("Normalized Data")
plt.show()
```

```python
ones = np.ones((X.shape[0],1))  # adding x0 to all the rows
X_ = np.hstack((ones,X))
print(X_.shape)
print(X_[:5,:])
```

```python
def predict(X,theta):
    return np.dot(X,theta)
```

```python
def getThetaClosedForm(X,Y):
    Y = np.mat(Y)
    
    firstpart = np.dot(X.T,X)
    secondpart = np.dot(X.T,Y)
    
    theta = np.linalg.pinv(firstpart)*secondpart
    return theta
```

```python
theta = getThetaClosedForm(X_,Y)
print(theta)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X,predict(X_,theta),color='red',label='prediction')

plt.legend()
plt.show()
```
