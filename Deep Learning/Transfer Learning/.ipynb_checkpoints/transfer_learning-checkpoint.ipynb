```python
import numpy as np
import os
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.optimizers import Adam
from keras.layers import Dense,Activation,Flatten,Input,Dropout,GlobalAveragePooling2D
from keras.models import Model
from keras.utils import np_utils
import random

import matplotlib.pyplot as plt
```

```python
dirs = os.listdir("/home/vasu/all_projects/CB/machine-learning-aug-18/class_16/data/")
```

```python
print dirs
```

```python
folder_path = "/home/vasu/all_projects/CB/machine-learning-aug-18/class_16/data/"
```

```python
image_data = []
```

```python
labels = []
```

```python
label_dict = {'cats':0,'dogs':1,'horses':2,'Humans':3}
```

```python
for ix in dirs:
    path = folder_path + ix + '/'
    img_data = os.listdir(path)
    for im in img_data:
        img = image.load_img(path+im,target_size=(224,224))
        img_array = image.img_to_array(img)
        image_data.append(img_array)
        labels.append(label_dict[ix])
```

```python
print len(image_data), len(labels)
```

```python
combined = list(zip(image_data, labels))
random.shuffle(combined)

image_data[:], labels[:] = zip(*combined)
```

```python
X_train = np.array(image_data)
Y_train = np.array(labels)

print X_train.shape,Y_train.shape
```

```python
Y_train = np_utils.to_categorical(Y_train)
print X_train.shape,Y_train.shape
```

```python
# predicting labels using resnets
model_test = ResNet50(include_top=True,weights='imagenet',input_shape=(224,224,3))

img_ = image.load_img("download.jpeg",target_size=(224,224))
img_array_ = image.img_to_array(img_)
x = np.expand_dims(img_array_,axis=0)
print x.shape
y = model_test.predict(x)
np.argmax(y)
```

```python
model = ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3))
```

```python
model.summary()
```

```python
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256,activation='relu')(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(4,activation='softmax')(d1)

model_new = Model(inputs=model.input, outputs= fc2)
model_new.summary()
```

```python
adam = Adam(lr=0.00003)
model_new.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
for ix in range(len(model_new.layers)):
    print ix, model_new.layers[ix]
```

```python
for ix in range(169):
    model_new.layers[ix].trainable = False
    
model_new.summary()
```

```python
hist = model_new.fit(X_train,Y_train,
                    shuffle = True,
                    batch_size = 16,
                    epochs = 5,
                    validation_split=0.20
                    )
```

```python
plt.figure(0)
plt.plot(hist.history['acc'],'b')
plt.plot(hist.history['val_acc'],'g')
plt.plot(hist.history['loss'],'black')
plt.plot(hist.history['val_loss'],'red')
plt.show()
```
