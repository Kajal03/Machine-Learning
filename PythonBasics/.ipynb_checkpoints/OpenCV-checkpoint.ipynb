```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('bulbasaur.png')
```

```python
plt.imshow(img)
```

```python
print(img.shape)
```

```python
rgb_img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(rgb_img)
```

```python
gray_img = cv2.cvtColor(rgb_img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img)
```

```python
crop_img = cv2.resize(rgb_img,(100,100))
```

```python
plt.imshow(crop_img)
```

```python
cropped = rgb_img[25:100,:70,:]
```

```python
plt.imshow(cropped)
```
