# Bar Graphs

```python
numbers = np.random.randint(0,10,5)
numbers2 = np.random.randint(0,10,5)
indices = np.arange(5)

print(numbers)
print(indices)
```

```python
plt.bar(indices,numbers,color='orange',label='Gold')
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn")
plt.figure(figsize=(7,8))
plt.bar(indices*2,numbers,width=0.5,color='gold',label='Current Year',tick_label=["Gold","Silver","Platinum","Brass","Copper"])
plt.bar((indices+0.3)*2,numbers2,width=0.5,color='silver',label='Next Year')
plt.legend()
plt.xlim(-1,10)
plt.ylim(0,10)
plt.show()
```

```python
plt.style.use("dark_background")
labels = ['Algebra', 'Calculus', 'Coding', 'Algorithms']
values = [90,80,54,40]

plt.pie(values,labels=labels, radius=1, explode=(0,0.1,0,0), shadow='True',autopct='%1.1f%%')
plt.show()
```
