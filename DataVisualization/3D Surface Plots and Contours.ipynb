```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [1,2,3]
b = [4,5,6,7]

a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
a = [-3,-2,-1,0,1,2,3]
b=a

a,b = np.meshgrid(a,b)

print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='coolwarm')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='coolwarm')
plt.show()
```
