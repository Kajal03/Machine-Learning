```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
dfX = pd.read_csv("xtrain.csv")
dfY = pd.read_csv("ytrain.csv")


x = dfX.values
y = dfY.values
```

```python
X = (x-x.mean())/x.std()
Y = y
```

```python
plt.scatter(X,Y,color="blue")
plt.scatter(x,y,color="orange")
plt.show()
```

```python
def hypothesis(theta,X):
    return theta[0] + theta[1]*X

def error(theta,X,Y):
    err=0
    m = X.shape[0]
    for i in range(m):
        err += (hypothesis(theta,X[i]) - Y[i])**2
    return err

def summation_hyp(theta,X,Y):
    grad = np.zeros((2,))
    for i in range(X.shape[0]):
        hyp = hypothesis(theta,X[i])
        grad[0]+= hyp-Y[i]
        grad[1]+= (hyp-Y[i])*X[i]
    return grad

def gradient(X,Y):
    theta=np.array([-3.0,-0.3])
    theta[0]=theta[1]=0
    learning_rate=0.005
    error_list = []
    theta_list = []
    m=X.shape[0]
    max_itr = 200
    for i in range(max_itr):
        grad = summation_hyp(theta,X,Y)
        er = error(theta,X,Y)
        error_list.append(er)
        
        
        theta[0]=theta[0] - (learning_rate*grad[0])
        theta[1]=theta[1] - (learning_rate*grad[1])        
        theta_list.append((theta[0],theta[1]))
        
    return theta, error_list, theta_list


final_theta, error, theta_list=gradient(X,Y)
xtest =np.linspace(-2,6,10)
plt.plot(error)
plt.show()

plt.plot(xtest,hypothesis(final_theta,xtest),label="Predicted Line")
plt.scatter(X,Y,color="orange",label="Training Data")
plt.legend()
plt.show()
    
```
