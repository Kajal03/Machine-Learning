```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfX = pd.read_csv('xtrain.csv')
dfY = pd.read_csv('ytrain.csv')

x = dfX.values
y = dfY.values

print(x.shape)
print(y.shape)
```

```python
X = (x-x.mean())/x.std()
Y = y
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
def hyp(theta,x):
    return (theta[0] + theta[1]*x)

def gradient(x,y,theta):
    return abs(hyp(theta,x)-y), (hyp(theta,x)-y)

def stochastic(x,y):
    theta =np.zeros((2,))
    #theta[0] = theta[1] = 0
    learning_rate = 0.1
    error_list =[]
    for i in range(x.size):
        err, grad = gradient(x[i],y[i],theta)
        error_list.append(err)
        theta[0] -= learning_rate*grad
        theta[1] -= learning_rate*grad*x[i]
    return theta,error_list
```

```python
theta, final_error = stochastic(X,Y)
```

```python
plt.plot(final_error)
```

```python
plt.plot(X,hyp(theta,X),label="Predicted Line")
plt.scatter(X,Y,color="orange",label="Training Data")
plt.legend()
plt.show()
```
